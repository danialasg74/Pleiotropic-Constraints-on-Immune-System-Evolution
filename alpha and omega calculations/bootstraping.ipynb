```python
import os
import numpy as np
import pandas as pd

import os
import pandas as pd

#import popfly
os.chdir('/Users/danial/Documents/postdoc_research/Domains/My_approach/seq_3_groups/main/updated/AllDomains/sixspeices/alpha_estim/DEST/degenotate')
popflydata  = pd.read_csv('popflydata.csv', index_col=0)

#import 3 classes of gene ids
os.chdir('classes')
pleio = pd.read_csv('Pleiotropic.csv', index_col=0)
immun = pd.read_csv('Non_Pleiotropic_immune.csv', index_col=0)
devel = pd.read_csv('Non_Pleiotropic_Dev.csv', index_col=0)

pleio = pleio.iloc[:, 0]
immun = immun.iloc[:, 0]
devel = devel.iloc[:, 0]


#keep genes for which we have divergence data
genes_in_popfly = set(popflydata["Name"])
pleio = pleio[pleio.isin(genes_in_popfly)]
immun = immun[immun.isin(genes_in_popfly)]
devel = devel[devel.isin(genes_in_popfly)]




###################### 100 bootstrap files for each gene category ######################

########################################################################################


import os
import numpy as np
import pandas as pd

#set seed
np.random.seed(1)

#number of bootstrap replicates
B = 100

os.chdir('/Users/danial/Documents/postdoc_research/Domains/My_approach/seq_3_groups/main/updated/AllDomains/sixspeices/alpha_estim/DEST/degenotate/bootstrap_genes2/pleio')
for b in range(1, B + 1):
    
    pleio_bs = np.random.choice(pleio, size=len(pleio), replace=True)
    pleio_bs_df = pd.DataFrame(pleio_bs, columns=["pleio"])
    
    #save files
    pleio_bs_df.to_csv(f"pleio_bootstrap_{b}.csv", index=False)
   
   
os.chdir('/Users/danial/Documents/postdoc_research/Domains/My_approach/seq_3_groups/main/updated/AllDomains/sixspeices/alpha_estim/DEST/degenotate/bootstrap_genes2/immun')
for b in range(1, B + 1):
    
    immun_bs = np.random.choice(immun, size=len(immun), replace=True)
    immun_bs_df = pd.DataFrame(immun_bs, columns=["immun"])
    
    #save files
    immun_bs_df.to_csv(f"immun_bootstrap_{b}.csv", index=False)
   
   
os.chdir('/Users/danial/Documents/postdoc_research/Domains/My_approach/seq_3_groups/main/updated/AllDomains/sixspeices/alpha_estim/DEST/degenotate/bootstrap_genes2/devel')
for b in range(1, B + 1):
    
    devel_bs = np.random.choice(devel, size=len(devel), replace=True)
    devel_bs_df = pd.DataFrame(devel_bs, columns=["devel"])
    
    #save files
    devel_bs_df.to_csv(f"devel_bootstrap_{b}.csv", index=False)



```
