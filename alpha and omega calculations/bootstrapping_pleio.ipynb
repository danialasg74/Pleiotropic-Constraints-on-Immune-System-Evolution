```python
import os
import numpy as np
import pandas as pd


################## round function ##################
#Convert a projected (fractional) Site Frequency Spectrum (SFS) into
#integer counts while preserving the total number of sites.

def round_sfs_preserve_total(sfs_float: np.ndarray) -> np.ndarray:
    x = np.asarray(sfs_float, dtype=float)
    x = np.maximum(x, 0.0)

    flo = np.floor(x).astype(int)
    remainder = x - flo

    total_target = int(round(x.sum()))
    current = flo.sum()
    need = total_target - current

    if need > 0:
        order = np.argsort(-remainder)  # largest remainder first
        flo[order[:need]] += 1

    return flo

################## import the MUN pop ##################

os.chdir('/Users/danial/Documents/postdoc_research/Domains/My_approach/seq_3_groups/main/updated/AllDomains/sixspeices/alpha_estim/DEST/degenotate/pops_pergeneoutputs/DE_Bay_Mun_1_2015-06-01')
MUN_0 = np.load('gene_sfs_0D_DE_Bay_Mun_1_2015-06-01.npy', allow_pickle=True)
MUN_4 = np.load('gene_sfs_4D_DE_Bay_Mun_1_2015-06-01.npy', allow_pickle=True)

#extract values
MUN_0_item = MUN_0.item()
MUN_4_item = MUN_4.item()

################## run the bootstrap ##################

n_boot = 100
n_chrom = 20

bootstrap_dir = "../bootstrap_genes/pleio"

output_dir = "../output"

# Make sure output directory exists
os.makedirs(output_dir, exist_ok=True)

# ---- LOOP ----
for b in range(1, n_boot + 1):

    bootstrap_file = os.path.join(bootstrap_dir, f"pleio_bootstrap_{b}.csv")

    if not os.path.exists(bootstrap_file):
        print(f"Skipping missing file: {bootstrap_file}")
        continue

    pleio = pd.read_csv(bootstrap_file)

    arrays0 = [
        MUN_0_item[idx]
        for idx in pleio['pleio']
        if idx in MUN_0_item
    ]

    arrays4 = [
        MUN_4_item[idx]
        for idx in pleio['pleio']
        if idx in MUN_4_item
    ]

    # Safety check
    if len(arrays0) == 0 or len(arrays4) == 0:
        print(f"Bootstrap {b} has no matching genes. Skipping.")
        continue

    total_0 = np.sum(arrays0, axis=0)
    total_4 = np.sum(arrays4, axis=0)

    total_0_int = round_sfs_preserve_total(total_0)
    total_4_int = round_sfs_preserve_total(total_4)

    out_file = os.path.join(output_dir, f"pleio_bootstrap_{b}.sfs")

    with open(out_file, "w") as f:
        f.write(f"{n_chrom}\n")
        f.write(" ".join(str(x) for x in total_0_int) + "\n")
        f.write(" ".join(str(x) for x in total_4_int) + "\n")

    print(f"Wrote: pleio_bootstrap_{b}.sfs")

print("All bootstraps completed.")
```
